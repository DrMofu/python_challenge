问len(a[30])=?

1 -> 1个1 -> 11 -> 2个1 -> 21 -> 1个2,1个1 -> 1211 -> 1个1,1个2,2个1 -> 111221 ->...

```python
string = '1'
for i in range(30):
    counter = [[string[0],0]]
    # 统计
    for char in string:
        if char==counter[-1][0]:
            counter[-1][1] += 1
        else:
            counter.append([char,1])
    # 生成新字符串
    string = ''
    for item in counter:
        string += item[0] + str(item[1])
print(len(string))
```
