```python
smallWordList = list('abcdefghijklmnopqrstuvwxyz')
bigWordList = [letter.upper() for letter in smallWordList]
```

```python
Sentence = ''
CharList = ''
with open('material/3/Sentence.txt', 'r') as f:
    Sentence = f.readlines()
for i in range(len(Sentence)):
    CharList += Sentence[i].strip()
```

```python
result = ''
preSmallLetter = '' # 符合恰好只有前三个字母是大写字母的小写字母。
BigCounter = 0
for i in range(len(CharList)):
    currentLetter  = CharList[i]
    if currentLetter in bigWordList:
        BigCounter += 1
    else:
        if BigCounter == 3:
            result += preSmallLetter
            preSmallLetter = currentLetter
        else:
            preSmallLetter = ''        
        BigCounter = 0
```

### level 4: follow the chain
http://www.pythonchallenge.com/pc/def/linkedlist.php

考察：爬虫基本原理

```python
import requests
from bs4 import BeautifulSoup
```

```python
# 示例
baseURL = 'http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing='
response = requests.get(baseURL+'12345')
response.encoding = 'utf-8'
soup = BeautifulSoup(response.text,'lxml')
soup.text.split()
```

```python
# 简易爬虫。不使用vpn可能会速度极慢
def getNext(currentNum):
    for i in range(100):
        try:
            response = requests.get(baseURL+currentNum)
            response.encoding = 'utf-8'
            soup = BeautifulSoup(response.text,'lxml')
            nextURL = soup.text.split()[-1]
            if nextURL == 'EOF':
                continue
            int(nextURL)
            currentNum = nextURL
            print(soup.text)
        except:        
            print(soup.text)
            break
```

已知
a = [1, 11, 21, 1211, 111221,...]

问len(a[30])=?

1 -> 1个1 -> 11 -> 2个1 -> 21 -> 1个2,1个1 -> 1211 -> 1个1,1个2,2个1 -> 111221 ->...

```python
string = '1'
for i in range(30):
    counter = [[string[0],0]]
    # 统计
    for char in string:
        if char==counter[-1][0]:
            counter[-1][1] += 1
        else:
            counter.append([char,1])
    # 生成新字符串
    string = ''
    for item in counter:
        string += item[0] + str(item[1])
print(len(string))
```

evil1.jpg中，纸牌被分为了5类。暗示我们把文件分为5类

于是读入evil2.gfx文件，将文件分为5类（分类方法同level 11）

使用pillow的Image.open读取被分割后的图片文件，且自动判别图片文件的格式，并保存

```python
with open("./material/12/evil2.gfx", "rb") as f:
    data = f.read()
```

```python
from PIL import Image
```

```python
for i in range(5):
    piece = data[i::5]
    im = Image.open(io.BytesIO(piece))
    with open("./material/12/%d.%s" % (i, im.format.lower()), "wb") as f:
        f.write(piece)
```
